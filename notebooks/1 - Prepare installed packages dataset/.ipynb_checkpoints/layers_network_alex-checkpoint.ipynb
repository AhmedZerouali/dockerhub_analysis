### Let's identify the images that are required to build other images


```python
# identify unique images
list_images = (layers
                 .drop(['kind','layers','added','depends'], axis=1)
                 .rename(columns={'image':'requires','hash':'depends'})
                 .set_index('requires')
                 .drop_duplicates()
                 .reset_index())
```

```python
# Merge hash with depends
layers = (layers
          .drop(['kind','layers','added'], axis=1)
          .dropna()
          .merge(list_images,
                 on='depends',
                 how='left')
         )
```
